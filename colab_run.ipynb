# PDF to DB - Google Colab

Menjalankan aplikasi PDF to DB di Google Colab dengan ngrok untuk akses dari internet.

**Langkah:**
1. Daftar gratis di [ngrok](https://ngrok.com) → [Dashboard](https://dashboard.ngrok.com/get-started/your-authtoken) → Copy auth token
2. Ganti `YOUR_NGROK_TOKEN` di cell pertama dengan token Anda
3. Jalankan semua cell (Runtime → Run all)

## Opsi: Satu Cell (All-in-One)

```python
# Copy-paste dan jalankan satu cell ini untuk setup lengkap
import os
os.environ["NGROK_AUTH_TOKEN"] = "YOUR_NGROK_TOKEN"  # Ganti dengan token ngrok

!git clone -q https://github.com/Herutriana44/pdf_to_db.git
%cd pdf_to_db
!pip install -q -r requirements.txt pyngrok

import threading, time
def run(): from app import app; app.run(host="0.0.0.0", port=5000, debug=False, use_reloader=False)
threading.Thread(target=run, daemon=True).start()
time.sleep(4)

from pyngrok import ngrok
ngrok.set_auth_token(os.environ.get("NGROK_AUTH_TOKEN", ""))
url = ngrok.connect(5000)
print(f"\n✅ Akses PDF to DB: {url}")
```

```python
import threading
import time

def run_flask():
    from app import app
    app.run(host="0.0.0.0", port=5000, debug=False, use_reloader=False)

thread = threading.Thread(target=run_flask, daemon=True)
thread.start()
time.sleep(4)
print("Flask app berjalan di port 5000")
```

```python
from pyngrok import ngrok

if os.environ.get("NGROK_AUTH_TOKEN"):
    ngrok.set_auth_token(os.environ["NGROK_AUTH_TOKEN"])

public_url = ngrok.connect(5000)
print(f"\nAkses web PDF to DB: {public_url}")
```
